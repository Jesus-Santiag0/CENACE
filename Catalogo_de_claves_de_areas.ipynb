<h4>Algoritmo realizado con el fin de limpiar y conseguir las coordenadas de las plantas de generación de energía del CENACE</>


```python
import pandas as pd #Importamos a Pandas
from pathlib import Path
```

```python
# Leemos el archivo
df=pd.read_csv('G:/Datos/Mapa México/MUN/AGEEML_2026113136603_utf.csv')
df.info()
```

```python
existe= "Acapulco" in df["NOM_LOC"].values
print(existe)
```

```python
# Eliminación de columnas que no son de utilidad
df_colums=df.drop(["CVEGEO","Estatus","NOM_ABR","AMBITO","LATITUD","LONGITUD","ALTITUD","CVE_CARTA","POB_TOTAL","POB_MASCULINA","POB_FEMENINA","TOTAL DE VIVIENDAS HABITADAS"],axis=1)
df_colums.info()
```

```python
# Filtro para dejar solamente a los municipios
df_limpio=df_colums.drop_duplicates(subset=['CVE_ENT','CVE_MUN','CVE_LOC'])
df_limpio.info()
df_limpio.head()
```

```python
# Guardamos el archivo
df_limpio.to_excel('G:/Datos_limpios/Mapa Mexico/Coordenadas_de_Municipios.xlsx', index=False)
```
