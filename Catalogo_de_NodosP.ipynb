<h4>Algoritmo realizando con el fin de limpiar los archivos .csv de Catálogos de Nodos P</>

```python
from pathlib import Path
import pandas as pd
```

```python
ano=2024
# Escribir las fechas de publicación del catálogo de Nodos con el siguiente formato: AAAA-MM-DD
ano2025=['2025-01-21','2025-02-19','2025-03-21','2025-04-17','2025-05-27','2025-06-18','2025-07-31','2025-08-20','2025-09-26','2025-10-17','2025-11-15']
ano2024=['2024-01-24','2024-02-20','2024-03-22','2024-04-25','2024-05-18','2024-06-26','2024-07-21','2024-08-29','2024-09-21','2024-10-18','2024-11-20','2024-12-17']
ano2026=['2026-01-24']
```

```python
directorio_origen = Path(f'C:/Users/Santi/OneDrive - Instituto Politecnico Nacional/Documents/CIC/Datos/Catalogos de NodosP/{ano}')

# Ciclo for que recorre las fechas
# !Cambiar el año
for fecha in ano2024:
	# Busca el archivo que contiene esta fecha en su nombre
	archivos = list(directorio_origen.glob(f'*v{fecha}*.xlsx'))
	if archivos:
		ruta_origen = archivos[0]
		df = pd.read_excel(ruta_origen, skiprows=1)
		df.to_excel(f'./Datos_limpios/Catalogos de NodosP/{ano}/Catálogo NodosP Sistema Eléctrico Nacional (v{fecha}).xlsx', index=False)
		print(f"Procesado: {fecha}")
	else:
		print(f"No encontrado: {fecha}")
```

```python
# Crea una lista que contiene DataFrames
df_n=[]
directorio_procesado = Path(f'./Datos_limpios/Catalogos de NodosP/{ano}')

# !Cambiar el año
for fecha in ano2024:
	# Busca el archivo que contiene esta fecha en su nombre
	archivos = list(directorio_procesado.glob(f'*v{fecha}*.xlsx'))
	if archivos:
		df = pd.read_excel(archivos[0])
		df_n.append(df)

# Concatena los registros
df_completo=pd.concat(df_n,ignore_index=True)
```

```python
# Elmina filas duplicadas por la Clave del Nodod
# df_completop=df_completo.drop_duplicates() #?2624
df_completop=df_completo.drop_duplicates(subset=['CLAVE','SISTEMA']) #? 2593
# ? Este cambio indica que 31 registros Cambiaron valores en algunas de sus columnas
df_completop.info()

df_completop.to_excel(f'./Datos_limpios/Catalogos de NodosP/{ano}/Catálogo NodosP Sistema Eléctrico Nacional (v{ano}).xlsx', index=False)
```
