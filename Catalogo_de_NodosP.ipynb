<h4>Algoritmo realizando con el fin de limpiar los archivos .csv de Catálogos de Nodos P</>

```python
from pathlib import Path
import pandas as pd
```

```python
ano=2025
directorio_procesado = Path(f'G:/Datos_limpios/Catalogos de NodosP/{ano}')
directorio_procesado.mkdir(parents=True,exist_ok=True)
# Algoritmo para obtener las fechas automáticamente
# + En dado caso de que en un futura necesitamos usar el "código pendiente" podemos guardar las fechas en un archivo .txt
directorio_origen = Path(f'G:/Datos/Catalogos de NodosP/{ano}')
archivos=list(directorio_origen.glob('*.xlsx*'))
fechas=[]
for archivo in archivos:
	ini=archivo.name.find('v')
	ini=ini+1
	fin=ini+10
	cadena=archivo.name[ini:fin]
	fechas.append(cadena)
	# Imprimimos para ver si están bien las fechas
	# print(cadena)
	df = pd.read_excel(archivo, skiprows=1)
	df.to_excel(f'G:/Datos_limpios/Catalogos de NodosP/{ano}/{archivo.name}', index=False)
	print(f"Procesado: {archivo.name}")
```

```python
# Crea una lista que contiene DataFrames
directorio_procesado = Path(f'G:/Datos_limpios/Catalogos de NodosP/{ano}')
archivos=list(directorio_procesado.glob('*.xlsx*'))
df_n=[]
for archivo in archivos:
	df=pd.read_excel(archivo)
	df_n.append(df)

# Concatena los registros
df_completo=pd.concat(df_n,ignore_index=True)
```

```python
# Elmina filas duplicadas por la Clave del Nodod
# df_completop=df_completo.drop_duplicates() #?2624
df_completop=df_completo.drop_duplicates(subset=['CLAVE','SISTEMA']) #? 2593
# ? Este cambio indica que 31 registros Cambiaron valores en algunas de sus columnas
df_completop.info()

df_completop.to_excel(f'G:/Datos_limpios/Catalogos de NodosP/{ano}/Catálogo NodosP Sistema Eléctrico Nacional (v{ano}).xlsx', index=False)
```

```python
# Unión de todos los archivos
# !2025-2020
df_1=pd.read_excel('G:/Datos_limpios/Catalogos de NodosP/2025/Catálogo NodosP Sistema Eléctrico Nacional (v2025).xlsx')
df_2=pd.read_excel('G:/Datos_limpios/Catalogos de NodosP/2024/Catálogo NodosP Sistema Eléctrico Nacional (v2024).xlsx')
df_3=pd.read_excel('G:/Datos_limpios/Catalogos de NodosP/2023/Catálogo NodosP Sistema Eléctrico Nacional (v2023).xlsx')
df_4=pd.read_excel('G:/Datos_limpios/Catalogos de NodosP/2022/Catálogo NodosP Sistema Eléctrico Nacional (v2022).xlsx')
df_5=pd.read_excel('G:/Datos_limpios/Catalogos de NodosP/2021/Catálogo NodosP Sistema Eléctrico Nacional (v2021).xlsx')
df_6=pd.read_excel('G:/Datos_limpios/Catalogos de NodosP/2020/Catálogo NodosP Sistema Eléctrico Nacional (v2020).xlsx')

df_catalog_completo=pd.concat([df_1,df_2,df_3,df_4,df_5,df_6],axis='rows')
# Quitamos los duplicados
df_catalog_completop=df_catalog_completo.drop_duplicates(subset=['CLAVE','SISTEMA'])
df_catalog_completop.to_excel('G:/Datos_limpios/Catalogos de NodosP/Catálogo NodosP Sistema Eléctrico Nacional (vCOMPLETO).xlsx',index=False)
```
