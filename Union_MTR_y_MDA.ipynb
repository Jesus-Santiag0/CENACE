<h4>Algoritmo para juntar columnas del precio MTR y MDA del CENACE</>
<h5>De los archivos .csv previamente transformados del mismo año</>

```python
import pandas as pd #importamos pandas
```

```python
# Leemos los archivos Nacionales por año
ano=2025 #cambiar el valor
df_Nacional_MTR=pd.read_csv(f'./Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR v{ano}.csv')
df_Nacional_MDA=pd.read_csv(f'./Datos_limpios/Nacional/MDA/PreciosMargLocales NACIONAL MDA v{ano}.csv')
```

```python
# Coborramos de que estén correctos
df_Nacional_MTR.info()
df_Nacional_MDA.info()
```

```python
# Eliminamos las columnas no deseadas
df_Nacional_MDA.drop(columns=['Unnamed: 0'], inplace=True)
df_Nacional_MTR.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Colocamos nuevos nombres representativos a las columnas
df_Nacional_MTR.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh) MTR','Componente de energia ($/MWh) MTR','Componente de perdidas ($/MWh) MTR','Componente de congestion ($/MWh) MTR']
df_Nacional_MDA.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh) MDA','Componente de energia ($/MWh) MDA','Componente de perdidas ($/MWh) MDA','Componente de congestion ($/MWh) MDA']
```

```python
# Coborramos de que estén correctos
df_Nacional_MTR.info()
df_Nacional_MDA.info()
```

```python
# Combinamos las columnas de los df usando las columnas en común que son: FECHA, HORA y CLAVE DEL NODO
df_union=pd.merge(df_Nacional_MTR,df_Nacional_MDA,on=['Fecha','Hora','Clave del nodo'])
# Verificamos los primeros registros
df_union.info()
df_union.head()
```

```python
# Guardamos el dataframe en un archivo .csv
df_union.to_csv(f'./Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA v{ano}.csv',index=False)
```
