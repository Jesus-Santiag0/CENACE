<h4>Algoritmo para juntar columnas del precio MTR y MDA del CENACE</>
<h5>De los archivos .csv previamente transformados del mismo año</>

```python
import pandas as pd #importamos pandas
```

```python
# Leemos los archivos Nacionales por año
ano=2018 #cambiar el valor
df_Nacional_MTR=pd.read_csv(f'G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR v{ano}.csv')
df_Nacional_MDA=pd.read_csv(f'G:/Datos_limpios/Nacional/MDA/PreciosMargLocales NACIONAL MDA v{ano}.csv')
```

```python
# Colocamos nuevos nombres representativos a las columnas
df_Nacional_MTR.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh) MTR','Componente de energia ($/MWh) MTR','Componente de perdidas ($/MWh) MTR','Componente de congestion ($/MWh) MTR']
df_Nacional_MDA.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh) MDA','Componente de energia ($/MWh) MDA','Componente de perdidas ($/MWh) MDA','Componente de congestion ($/MWh) MDA']
```

```python
# Coborramos de que estén correctos
df_Nacional_MTR.info()
df_Nacional_MDA.info()
```

```python
# Registros en MTR que no están en MDA
solo_en_MTR = df_Nacional_MTR.merge(df_Nacional_MDA, on=['Fecha','Hora','Clave del nodo'], how='left', indicator=True)
solo_en_MTR = solo_en_MTR[solo_en_MTR['_merge'] == 'left_only']
print(solo_en_MTR[['Fecha','Hora','Clave del nodo']])
print('--------')
# Registros en MDA que no están en MTR
solo_en_MDA = df_Nacional_MDA.merge(df_Nacional_MTR, on=['Fecha','Hora','Clave del nodo'], how='left', indicator=True)
solo_en_MDA = solo_en_MDA[solo_en_MDA['_merge'] == 'left_only']
print(solo_en_MDA[['Fecha','Hora','Clave del nodo']])
# Guardar diferencias en un .txt
with open(f'G:/Datos_limpios/Nacional/MTR&MDA/diferencias_{ano}.txt', 'w', encoding='utf-8') as f:
    f.write('Registros en MTR que no están en MDA\n')
    f.write(solo_en_MTR[['Fecha','Hora','Clave del nodo']].to_string(index=False))
    f.write('\n\n--------\n\n')
    f.write('Registros en MDA que no están en MTR\n')
    f.write(solo_en_MDA[['Fecha','Hora','Clave del nodo']].to_string(index=False))
```

```python
# Combinamos las columnas de los df usando las columnas en común que son: FECHA, HORA y CLAVE DEL NODO
df_union=pd.merge(df_Nacional_MTR,df_Nacional_MDA,on=['Fecha','Hora','Clave del nodo'])
# Verificamos los primeros registros
df_union.info()
df_union.head()
```

```python
# Guardamos el dataframe en un archivo .csv
df_union.to_csv(f'G:/Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA v{ano}.csv',index=False)
```

```python
# Unión de todos los archivos
# !2025-2020
df_1=pd.read_csv(f'G:/Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA v2025.csv')
df_2=pd.read_csv(f'G:/Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA v2024.csv')
df_3=pd.read_csv(f'G:/Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA v2023.csv')
df_4=pd.read_csv(f'G:/Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA v2022.csv')
df_5=pd.read_csv(f'G:/Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA v2021.csv')
df_6=pd.read_csv(f'G:/Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA v2020.csv')
df_completo=pd.concat([df_1,df_2,df_3,df_4,df_5,df_6],axis='rows')
df_completo.to_csv('G:/Datos_limpios/Nacional/MTR&MDA/PreciosMargLocales NACIONAL MTR&MDA vCOMPLETO.csv',index=False,encoding='utf-8')
```
