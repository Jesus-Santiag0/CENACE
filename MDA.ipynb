```python
import pandas as pd #Importamos a Pandas
from pathlib import Path
```

```python
# Crea una lista con las rutas de los archivos
ano=2021
# BCA, SIN, BCS
sistema='BCA'
meses=['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']
rutas=[]
for i,mes in enumerate(meses):
	i=i+2
	for j in range(1,3):
		if(9< i <13):
			# name=f'PreciosMargLocales {sistema} MDA Mes {mes}0{j} v{ano} {i} 10_08 30 01.csv'
			# 2023
			name=f'PreciosMargLocales {sistema} MDA Mes {mes}0{j} v{ano} {i} 03_08 00 01.csv'
		elif(i==13):
			# Caso especial de diciembre
			# name=f'PreciosMargLocales {sistema} MDA Mes {mes}0{j} v{ano+1} 01 10_08 30 01.csv'
			# 2023
			name=f'PreciosMargLocales {sistema} MDA Mes {mes}0{j} v{ano+1} 01 03_08 00 01.csv'
		else:
			# name=f'PreciosMargLocales {sistema} MDA Mes {mes}0{j} v{ano} 0{i} 10_08 30 01.csv'
			# 2023
			name=f'PreciosMargLocales {sistema} MDA Mes {mes}0{j} v{ano} 0{i} 03_08 00 01.csv'
		rutas.append(name)
print(rutas)
```

```python
#! Esto queda pendiente de uso
# Escribir las fechas de publicación del catálogo de Nodos con el siguiente formato: AAAA-MM-DD
ano2025=['2025-01-21','2025-02-19','2025-03-21','2025-04-17','2025-05-27','2025-06-18','2025-07-31','2025-08-20','2025-09-26','2025-10-17','2025-11-15','2025-12-13']
ano2024=['2024-01-24','2024-02-20','2024-03-22','2024-04-25','2024-05-18','2024-06-26','2024-07-21','2024-08-29','2024-09-21','2024-10-18','2024-11-20','2024-12-17']
ano2026=['2026-01-24']
#fecha_atras<fecha<=fecha_adelante
# Se toma el catálogo cuya Fecha de actualización sea la más reciente antes del inicio del periodo de liquidación
#Hacer un filtro que seleccione el rango de fechas seleccionado y depués guardar
# fecha_ini=0
# fecha_fin=0
fechas_ini=[]
fechas_fin=[]
for i in range(0,12):
	if i==0:
		fecha_ini=ano2024[11]
		fecha_fin=ano2025[i]
		fechas_ini.append(fecha_ini)
		fechas_fin.append(fecha_fin)
	# Variar dependiendo el número de fechas que tenemos disponibles
	elif i==11:
		try:
			fecha_ini=ano2025[i]
			fecha_fin=ano2026[0]
			fechas_ini.append(fecha_ini)
			fechas_fin.append(fecha_fin)
		except:
			print("Aún no hay suficientes")
	else:
		try:
			fecha_ini=ano2025[i-1]
			fecha_fin=ano2025[i]
			fechas_ini.append(fecha_ini)
			fechas_fin.append(fecha_fin)
		except:
			print("Aún no hay suficientes")
print(fechas_ini)
	#Cocatenar todos los CSV
```

```python
# Código que lee los documentos .csv de MDA por mes y los concatena en 1
ruta=Path(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[0]} v{ano}.csv')
# Aseguramos de que existe la dirección
ruta.parent.mkdir(parents=True, exist_ok=True)
try:
	# Enero
	df_1=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[0]}',skiprows=7)
	df_2=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[1]}',skiprows=7)
	df_enero=pd.concat([df_1,df_2], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_enero.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[0]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Enero aún no están")
try:
	#Febrero
	df_3=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[2]}',skiprows=7)
	df_4=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[3]}',skiprows=7)
	df_febrero=pd.concat([df_3,df_4], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_febrero.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[1]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Febrero aún no están")
try:
	#Marzo
	df_5=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[4]}',skiprows=7)
	df_6=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[5]}',skiprows=7)
	df_marzo=pd.concat([df_5,df_6], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_marzo.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[2]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Marzo aún no están")
try:
	# Abril
	df_7=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[6]}',skiprows=7)
	df_8=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[7]}',skiprows=7)
	df_abril=pd.concat([df_7,df_8], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_abril.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[3]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Abril aún no están")
try:
	# Mayo
	df_9=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[8]}',skiprows=7)
	df_10=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[9]}',skiprows=7)
	df_mayo=pd.concat([df_9,df_10], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_mayo.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[4]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Mayo aún no están")
try:
	# Junio
	df_11=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[10]}',skiprows=7)
	df_12=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[11]}',skiprows=7)
	df_junio=pd.concat([df_11,df_12], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_junio.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[5]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Junio aún no están")
try:
	# Julio
	df_13=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[12]}',skiprows=7)
	df_14=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[13]}',skiprows=7)
	df_julio=pd.concat([df_13,df_14], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_julio.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[6]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Julio aún no están")
try:
	# Agosto
	df_15=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[14]}',skiprows=7)
	df_16=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[15]}',skiprows=7)
	df_agosto=pd.concat([df_15,df_16], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_agosto.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[7]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Agosto aún no están")
try:
	# Septiembre
	df_17=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[16]}',skiprows=7)
	df_18=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[17]}',skiprows=7)
	df_septiembre=pd.concat([df_17,df_18], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_septiembre.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[8]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Septiembre aún no están")
try:
	# Octubre
	df_19=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[18]}',skiprows=7)
	df_20=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[19]}',skiprows=7)
	df_octubre=pd.concat([df_19,df_20], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_octubre.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[9]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Octubre aún no están")
try:
	# Noviembre
	df_21=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[20]}',skiprows=7)
	df_22=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[21]}',skiprows=7)
	df_noviembre=pd.concat([df_21,df_22], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_noviembre.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[10]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Noviembre aún no están")
try:
	# Diciembre
	df_23=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[22]}',skiprows=7)
	df_24=pd.read_csv(f'G:/Datos/Por Nodo/{sistema}/MDA/{ano}/{rutas[23]}',skiprows=7)
	df_diciembre=pd.concat([df_23,df_24], axis='rows') #Concatena los dos archivos por mes
	# Guardamos los DataFrames concatenados
	df_diciembre.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA Mes {meses[11]} v{ano}.csv', index=False, encoding='utf-8')
except:
	print("Los archivos de Diciembre aún no están")
```

```python
#Código que concatena todos los archivos limpios por mes

#!Colocar los archivos que ya tienen DataFrame
df_completo=pd.concat([df_enero,df_febrero,df_abril,df_mayo,df_junio,df_julio,df_marzo,df_agosto,df_septiembre,df_octubre,df_noviembre,df_diciembre], axis='rows')
# Guardamos el DataFrame completo
df_completo.to_csv(f'G:/Datos_limpios/{sistema}/MDA/{ano}/PreciosMargLocales {sistema} MDA v{ano}.csv', index=False, encoding='utf-8')
```

```python
# !Código pendiente de uso
# Dividir los datos de acuerdo al Catálogo de Nodos
df_2025=pd.read_csv('G:/MDA/2025/Completos/PreciosMargLocales SIN MDA v2025.csv')
df_2024=pd.read_csv('G:/MDA/2024/Completos/PreciosMargLocales SIN MDA v2024.csv')
# Unimos ambos años
df_union = pd.concat([df_2024, df_2025], ignore_index=True)
df_union['Fecha'] = pd.to_datetime(df_union['Fecha'])
```

```python
# !Código pendiente de uso
# Definimos fechas límite
for i,_ in enumerate(fechas_ini):
	fecha_inicio = pd.Timestamp(f'{fechas_ini[i]}')
	fecha_fin = pd.Timestamp(f'{fechas_fin[i]}')
	df_catalogo=df_union.query("Fecha > @fecha_inicio and Fecha <= @fecha_fin")
	print(fecha_inicio)
	print(fecha_fin)
	# Guardamos el DataFrame
	df_catalogo.to_csv(f'G:/MDA/{ano}/Catálogo_de_Nodos/PreciosMargLocales {sistema} MDA {fechas_ini[i]} {fechas_fin[i]}.csv', index=False, encoding='utf-8')
```
