```python
import pandas as pd #Importamos a Pandas
from pathlib import Path
```

```python
# Crea una lista con las rutas de los archivos
ano=2025
# BCA, SIN, BCS
sistema='BCS'
meses=['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']
rutas=[]
for i,mes in enumerate(meses):
	i=i+2
	for j in range(1,3):
		if(9< i <13):
			# name=f'PreciosMargLocales {sistema} MTR Mes {mes}0{j} v{ano} {i} 10_08 30 01.csv'
			# 2023
			name=f'PreciosMargLocales {sistema} MTR Mes {mes}0{j} v{ano}'
		elif(i==13):
			# Caso especial de diciembre
			# name=f'PreciosMargLocales {sistema} MTR Mes {mes}0{j} v{ano+1} 01 10_08 30 01.csv'
			# 2023
			name=f'PreciosMargLocales {sistema} MTR Mes {mes}0{j} v{ano+1}'
		else:
			# name=f'PreciosMargLocales {sistema} MTR Mes {mes}0{j} v{ano} 0{i} 10_08 30 01.csv'
			# 2023
			name=f'PreciosMargLocales {sistema} MTR Mes {mes}0{j} v{ano}'
		rutas.append(name)
print(rutas)
```

```python
#! Esto queda pendiente de uso
# Escribir las fechas de publicación del catálogo de Nodos con el siguiente formato: AAAA-MM-DD
ano2025=['2025-01-21','2025-02-19','2025-03-21','2025-04-17','2025-05-27','2025-06-18','2025-07-31','2025-08-20','2025-09-26','2025-10-17','2025-11-15','2025-12-13']
ano2024=['2024-01-24','2024-02-20','2024-03-22','2024-04-25','2024-05-18','2024-06-26','2024-07-21','2024-08-29','2024-09-21','2024-10-18','2024-11-20','2024-12-17']
ano2026=['2026-01-24']
#fecha_atras<fecha<=fecha_adelante
# Se toma el catálogo cuya Fecha de actualización sea la más reciente antes del inicio del periodo de liquidación
#Hacer un filtro que seleccione el rango de fechas seleccionado y depués guardar
# fecha_ini=0
# fecha_fin=0
fechas_ini=[]
fechas_fin=[]
for i in range(0,12):
	if i==0:
		fecha_ini=ano2024[11]
		fecha_fin=ano2025[i]
		fechas_ini.append(fecha_ini)
		fechas_fin.append(fecha_fin)
	# Variar dependiendo el número de fechas que tenemos disponibles
	elif i==11:
		try:
			fecha_ini=ano2025[i]
			fecha_fin=ano2026[0]
			fechas_ini.append(fecha_ini)
			fechas_fin.append(fecha_fin)
		except:
			print("Aún no hay suficientes")
	else:
		try:
			fecha_ini=ano2025[i-1]
			fecha_fin=ano2025[i]
			fechas_ini.append(fecha_ini)
			fechas_fin.append(fecha_fin)
		except:
			print("Aún no hay suficientes")
print(fechas_ini)
	#Cocatenar todos los CSV
```

```python
directorio_origen=Path(f'G:/Datos/Por Nodo/{sistema}/MTR/{ano}')
links=[]
for ruta in rutas:
	archivos=next(directorio_origen.glob(f'*{ruta}*.csv'))
	links.append(archivos)
print(links)
```

```python
#Código que concatena todos los archivos limpios por mes
dfs = [df_enero, df_febrero, df_marzo, df_abril, df_mayo, df_junio, df_julio, df_agosto, df_septiembre, df_octubre, df_noviembre, df_diciembre]
for df in dfs:
    df.columns = df.columns.str.strip()  # Esto elimina los espacios en blanco de las columnas

#!Colocar los archivos que ya tienen DataFrame
df_completo = pd.concat(dfs, axis='rows', ignore_index=True)
# Guardamos el DataFrame completo
df_completo.to_csv(f'G:/Datos_limpios/{sistema}/MTR/{ano}/PreciosMargLocales {sistema} MTR v{ano}.csv', index=False, encoding='utf-8')
```

```python
# Unión de todos los datos del MTR 2020-2025
df_mtr_2025=pd.read_csv('G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR v2025.csv')
df_mtr_2025.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh)','Componente de energia ($/MWh)','Componente de perdidas ($/MWh)','Componente de congestion ($/MWh)']
df_mtr_2024=pd.read_csv('G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR v2024.csv')
df_mtr_2024.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh)','Componente de energia ($/MWh)','Componente de perdidas ($/MWh)','Componente de congestion ($/MWh)']
df_mtr_2023=pd.read_csv('G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR v2023.csv')
df_mtr_2023.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh)','Componente de energia ($/MWh)','Componente de perdidas ($/MWh)','Componente de congestion ($/MWh)']
df_mtr_2022=pd.read_csv('G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR v2022.csv')
df_mtr_2022.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh)','Componente de energia ($/MWh)','Componente de perdidas ($/MWh)','Componente de congestion ($/MWh)']
df_mtr_2021=pd.read_csv('G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR v2021.csv')
df_mtr_2021.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh)','Componente de energia ($/MWh)','Componente de perdidas ($/MWh)','Componente de congestion ($/MWh)']
df_mtr_2020=pd.read_csv('G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR v2020.csv')
df_mtr_2020.columns=['Fecha','Hora','Clave del nodo','Precio marginal local ($/MWh)','Componente de energia ($/MWh)','Componente de perdidas ($/MWh)','Componente de congestion ($/MWh)']
df_completoMTR=pd.concat([df_mtr_2025,df_mtr_2024,df_mtr_2023,df_mtr_2022,df_mtr_2021,df_mtr_2020],axis='rows',ignore_index=True)
df_completoMTR.to_csv(f'G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR vCOMPLETO.csv',index=False, encoding='utf-8')
```

```python
df_completoMTR.info()
df_completoMTR.head()
```

```python
# Agrupa por Mes, Año y Clave del Nodo para sacar el promedio y así disminuir la cantidad de datos
# Primero convertimos la columna Fecha a datetime
df_completoMTR['Fecha'] = pd.to_datetime(df_completoMTR['Fecha'])

# Extrae mes y año
df_completoMTR['Año'] = df_completoMTR['Fecha'].dt.year
df_completoMTR['Mes'] = df_completoMTR['Fecha'].dt.month

# Agrupa por Clave del nodo y Año, y calcula el promedio
df_promedio_anual = df_completoMTR.groupby(['Clave del nodo', 'Año']).agg({
    'Precio marginal local ($/MWh)': 'mean',
    'Componente de energia ($/MWh)': 'mean',
    'Componente de perdidas ($/MWh)': 'mean',
    'Componente de congestion ($/MWh)': 'mean'
}).reset_index()
```

```python
# Agrupa por Clave del nodo, mes y año, y calcula el promedio
df_promedio_mensual = df_completoMTR.groupby(['Clave del nodo', 'Año','Mes']).agg({
    'Precio marginal local ($/MWh)': 'mean',
    'Componente de energia ($/MWh)': 'mean',
    'Componente de perdidas ($/MWh)': 'mean',
    'Componente de congestion ($/MWh)': 'mean'
}).reset_index()
df_promedio_mensual.info()
df_promedio_mensual.head()
```

```python
# Guardamos los datos agrupados por año
df_promedio_anual.to_csv('G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR vCOMPLETO_ANUAL.csv',index=False, encoding='utf-8')
```

```python
# Guardamos los datos agrupados por mes y año
df_promedio_mensual.to_csv('G:/Datos_limpios/Nacional/MTR/PreciosMargLocales NACIONAL MTR vCOMPLETO_MENSUAL.csv',index=False, encoding='utf-8')
```
